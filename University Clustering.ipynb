## K Means Cluster Creation

Now it is time to create the Cluster labels. As earlier stated, The labels will be dropped.

**Importing KMeans from SciKit Learn.**

```python
from sklearn.cluster import KMeans
```

**We will create an instance of a K Means model with 2 clusters because there are only two distinct types
of universities we want to cluster**

```python
kmeans = KMeans(n_clusters=2, init='k-means++',max_iter=300, random_state=0)
```

```python
X = df.drop(['Private',"Encoded"],axis=1)
```

```python
X.head(3)
```

**Fit the model to all the data except for the Private label.**

```python
kmeans.fit(X)
```

**cluster center vectors**

```python
kmeans.cluster_centers_
```

## Evaluation

There is no perfect way to evaluate clustering if you don't have the labels. However, we do have the labels, so we take advantage of this just to evaluate our clusters, keep in mind, we always won't have this luxury in the real world.

**We create a new column for df called 'Clusters', which is a 1 for a Private school, and a 0 for a public school.**

```python
df['Clusters'] = df['Private'].map({'Yes':1,'No':0})
```

def converter(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0

df['Cluster'] = df['Private'].apply(converter)

```python
df.head()
```

**And finally, we create a confusion matrix and classification report to see how well the the Kmeans clustering worked without being given any labels.**

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Clusters'],kmeans.labels_))
print(classification_report(df['Clusters'],kmeans.labels_))
```

**It is not so bad considering the Kmeans algorithm is purely using the features to cluster the universities into 2 distinct groups.**
